```python
def clean_to_minutes(time_text):
    if pd.isna(time_text):
        return 0
    time_text = str(time_text).strip().split()[-1]
    try:
        h, m, s = map(int, time_text.split(':'))
        return h * 60 + m
    except:
        return 0
```

```python
x= df['Distance(Km)'].tolist()
y = df['Delay_minutes'].tolist()
```

```python
mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)
```

```python
numerator = 0
denominator = 0
for i in range(len(x)):
    distance_diff = x[i] - mean_x
    delay_diff = y[i] - mean_y

    numerator += distance_diff * delay_diff
    denominator += distance_diff ** 2
m = numerator / denominator
c = mean_y - (m * mean_x)

print("--- MODEL TRAINING FINISHED ---")
print(f"Magic Slope (m): {m:.5f}")
print(f"Starting Point (c): {c:.2f}")
```

```python
new_train_distance = 1500
predicted_delay = (m * new_train_distance) + c
print("\n--- MAKING A NEW PREDICTION ---")
print(f"If a train travels {new_train_distance} km, our model guesses it will be late by: {predicted_delay:.2f} minutes!")
```
