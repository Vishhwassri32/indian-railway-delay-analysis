```python
# Top 10 longest routes
longest_routes = dt.groupby('Train_name')['Distance(Km)'].max().sort_values(ascending=False).head(10)
print(longest_routes)

# Plot
longest_routes.plot(kind='bar', title="Top 10 Longest Train Routes")
plt.ylabel("Distance (Km)")
plt.show()

# Average distance by train type
dist_by_type = dt.groupby('Train_name')['Distance(Km)'].mean().reset_index()
sns.barplot(data=dist_by_type, x='Train_name', y='Distance(Km)', palette='rainbow')
plt.title("Average Distance by Train Type")
plt.show()
```

```python
if {"Distance(Km)", "Delay_min"}.issubset(dt.columns):
    sns.regplot(x="Distance", y="Delay", data=dt, scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
    plt.title("Relationship Between Distance and Delay")
    plt.xlabel("Distance(Km)")
    plt.ylabel("Delay_min (minutes)")
    plt.show()
```

```python
# Select numeric columns only
num_cols = dt.select_dtypes(include=np.number)

# Correlation matrix
corr = num_cols.corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Railway Data")
plt.show()
```
